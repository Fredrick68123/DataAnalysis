```python
'''In the Analysis, One must Find what 

Questions can they Answear about the Data, 

how Usefull they are, 

and How to do it.'''

# Getting the the Prepared data file

import pandas as pd

data_file = pd.read_csv(r'stack-overflow-developer-survey-2022/survey_results_public.csv', index_col= 'ResponseId')
data_DF = pd.DataFrame(data_file)

pd.set_option('display.max_columns', 79)

#display(data_DF)

'''
Questions that I found : 1) are those who Learned programming Online more likely to be employed as a 

Developer than those who Learned at School or University?.

2) what is the most Popular field between employed Developers?.

3) how much do Education levels having jobs?.
'''





# Getting the needed Data for question #1

Q1_data = data_DF[['MainBranch', 'Employment', 'LearnCode']]



# Cleaning

Q1_data.dropna(inplace= True)

# Select those who are employed as Developers

developers_filter = Q1_data['MainBranch'].str.contains('I am a developer by profession')

Q1_data['MainBranch'] = Q1_data.loc[developers_filter, ['MainBranch']]



# Select those who are Employed

employed_filter = Q1_data['Employment'].str.contains('Employed')

Q1_data['Employment'] = Q1_data.loc[employed_filter, ['Employment']]



# Cleaning

Q1_data.dropna(inplace= True)



# Getting the number of those who have Learn't Online and not with school


online_learned_filter = Q1_data['LearnCode'].apply(lambda x: 'School (i.e., University, College, etc)' not in x)

online_learned_number = sum(Q1_data.loc[online_learned_filter, ['LearnCode']].value_counts())



# Getting the number of those who have Learn't with School and not Online

school_learned_filter = Q1_data['LearnCode'].apply(lambda x: 'Online Courses or Certification' not in list(x)) & Q1_data['LearnCode'].apply(lambda x: 'Other online resources (e.g., videos, blogs, forum)' not in x) & Q1_data['LearnCode'].apply(lambda x: 'Coding Bootcamp' not in x)

school_learned_number = sum(Q1_data.loc[school_learned_filter, ['LearnCode']].value_counts())

#print(school_learned_number, online_learned_number)



# Total number of developers

total_developer_number = sum(Q1_data['Employment'].value_counts())

def percentage(N, TN):
      print((N / TN) * 100)




# getting the Data for the second Question

Q2_data = data_DF[['MainBranch', 'Employment', 'DevType']]



# Cleaning

Q2_data.dropna(inplace= True)


# Select those who are employed as Developers

developers_filter_1 = Q2_data['MainBranch'].str.contains('I am a developer by profession')

Q2_data['MainBranch'] = Q2_data.loc[developers_filter_1, ['MainBranch']]



# Select those who are Employed

employed_filter_1 = Q2_data['Employment'].str.contains('Employed')

Q2_data['Employment'] = Q2_data.loc[employed_filter_1, ['Employment']]



# Cleaning

Q2_data.dropna(inplace= True)



Q2_data['DevType'].value_counts(sort= True)


# Backend

backend_filt =  Q2_data['DevType'].str.contains('back-end')

backend_number = sum(backend_filt)

# frontend

frontend_filt =  Q2_data['DevType'].str.contains('front-end')

frontend_number = sum(frontend_filt)

# fulstack

fullstack_filt =  Q2_data['DevType'].str.contains('full-stack')

fullstack_number = sum(fullstack_filt)

# Data Sience / Machine Learning

data_sience_machine_learning_filt =  Q2_data['DevType'].str.contains('Data scientist or machine learning specialist')

data_sience_machine_learning_number = sum(data_sience_machine_learning_filt)

# Data or business analyst

data_business_analyst_filt =  Q2_data['DevType'].str.contains('Data scientist or machine learning specialist')

data_business_analyst_number = sum(data_business_analyst_filt)

# Cloud infrastructure engineer

cloud_infrastructure_engineer_filt =  Q2_data['DevType'].str.contains('Cloud infrastructure engineer')

cloud_infrastructure_engineer_number = sum(cloud_infrastructure_engineer_filt)

# mobile

mobile_filt =  Q2_data['DevType'].str.contains('mobile')

mobile_number = sum(mobile_filt)

# DevOps specialist

devOps_specialist_filt =  Q2_data['DevType'].str.contains('DevOps specialist')

devOps_specialist_number = sum(devOps_specialist_filt)

# Database administrator

database_administrator_filt =  Q2_data['DevType'].str.contains('Database administrator')

database_administrator_number = sum(database_administrator_filt)



# getting the Data for the third Question

Q3_data = data_DF[['MainBranch', 'Employment', 'EdLevel', 'CompTotal']]



# Cleaning

Q3_data.dropna(inplace= True)


# Select those who are employed as Developers

developers_filter_2 = Q3_data['MainBranch'].str.contains('I am a developer by profession')

Q3_data['MainBranch'] = Q3_data.loc[developers_filter_2, ['MainBranch']]



# Select those who are Employed

employed_filter_2 = Q3_data['Employment'].str.contains('Employed')

Q3_data['Employment'] = Q3_data.loc[employed_filter_2, ['Employment']]



# Cleaning

Q3_data.dropna(inplace= True)


Q3_data['EdLevel'].value_counts()
```
